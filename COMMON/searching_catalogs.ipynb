## Searching catalogs

For data can be painful. We try to simplify the process by adding search function.

```python
import intake
import sys
sys.path.append("../")
import nereus as nr
```

```python
cat = intake.open_catalog("https://raw.githubusercontent.com/eerie-project/intake_catalogues/main/eerie.yaml")
```

```python
nr.search(cat, {"variable-long_names":"ice"})
```

```python
nr.search(cat['dkrz']['disk']['model-output']['icon-esm-er'], {"variable-long_names":"temperature"})
```

```python
data = cat['dkrz']['disk']['model-output']['icon-esm-er']['eerie-control-1950']['atmos']['gr025']['2d_monthly_mean'].to_dask()
```
